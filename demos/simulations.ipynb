# Tutorial for Simulations

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

First, import the simultions utilizing the import statement below:

```python
from mgcpy.benchmarks import simulations as sims
```

Simultions are randomly generated with an $x$ which is $(n\cdot d)$ and $y$ which is $(n\cdot 1)$ that have 2 required parameters: `num_samp` or the number of samples, and `num_dim` or the number of dimensions. Optional parameters can be set based on the documentation. Visualizations of the 4 simulations imported above are shown below with and without the noise. Each simulation is also shown in 2D.

## Linear Simulation

```python
x_normal, y_normal = sims.linear_sim(100, 1)
x_no_noise, y_no_noise = sims.linear_sim(1000, 1, noise=0)
```

```python
fig = plt.figure(figsize=(5, 4))
ax1 = fig.add_subplot(1, 1, 1)
ax1.scatter(x_normal, y_normal, c='b', label='Normal (with noise)')
ax1.scatter(x_no_noise, y_no_noise, c='r', label='No noise')
ax1.set_title('Linear', fontweight='bold')
ax1.set_xlabel('X Vector')
ax1.set_ylabel('Y Vector')
plt.legend(bbox_to_anchor=(1.3, 0.9), bbox_transform=plt.gcf().transFigure)
plt.show()
```
