# Tutorial for Simulations

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
import seaborn as sns; sns.set(color_codes=True)
```

First, import the simultions utilizing the import statement below:

```python
from mgcpy.benchmarks import simulations as sims
```

Simultions are randomly generated with an $x$ which is $(n\cdot d)$ and $y$ which is $(n\cdot 1)$ that have 2 required parameters: `num_samp` or the number of samples, and `num_dim` or the number of dimensions. Optional parameters can be set based on the documentation. Visualizations of the 4 simulations imported above are shown below with and without the noise. Each simulation is also shown in 2D.

For all the below simulations, $(X, Y) \overset{i.i.d}{\sim} F_{XY}=F_{Y|X}F_{X}$ where $X$ and $Y$ are independently identically distributed.

## Linear Simulation

Linear $(X,Y) \in \mathbb{R}^{p} \times \mathbb{R}$:

$$X \sim \mathcal{U}(-1,1)^{p}$$

$$Y=w^{T}X+\kappa \epsilon$$

```python
x_normal, y_normal = sims.linear_sim(100, 1)
x_no_noise, y_no_noise = sims.linear_sim(1000, 1, noise=0)
```

```python
fig = plt.figure(figsize=(5, 4))
ax1 = fig.add_subplot(1, 1, 1)
ax1.scatter(x_normal, y_normal, c='b', label='Normal (with noise)')
ax1.scatter(x_no_noise, y_no_noise, c='r', label='No noise')
ax1.set_title('Linear', fontweight='bold')
ax1.set_xlabel('X Vector')
ax1.set_ylabel('Y Vector')
plt.legend(bbox_to_anchor=(1.3, 0.9), bbox_transform=plt.gcf().transFigure)
plt.show()
```

## Joint-Normal Simulation

Joint normal $(X,Y) \in \mathbb{R}^{p} \times \mathbb{R}^{p}$: Let $\rho=1/2p$, $I_{p}$ be the identity matrix of size $p\times p$, $J_{p}$ be the matrix of ones of size $p\times p$, and $\Sigma =\begin{bmatrix}
    I_{p} & \rho J_{p} \\
    \rho J_{p} & (1+0.5\kappa)I_{p}\\
\end{bmatrix}$. Then

$$(X,Y) \sim \mathcal{N}(0,\Sigma)$$

```python
x_normal, y_normal = sims.joint_sim(100, 1)
x_no_noise, y_no_noise = sims.joint_sim(1000, 1, noise=0)
```

```python
fig = plt.figure(figsize=(5, 4))
ax1 = fig.add_subplot(1, 1, 1)
ax1.scatter(x_normal, y_normal, c='b', label='Normal (with noise)')
ax1.scatter(x_no_noise, y_no_noise, c='r', label='No noise')
ax1.set_title('Joint-Normal Simulation', fontweight='bold')
ax1.set_xlabel('X Vector')
ax1.set_ylabel('Y Vector')
plt.legend(bbox_to_anchor=(1.3, 0.9), bbox_transform=plt.gcf().transFigure)
plt.show()
```

## Step Function Simulation

Step Function $(X,Y) \in \mathbb{R}^{p} \times \mathbb{R}$:

$$X \sim \mathcal{U}(-1,1)^{p}$$

$$Y=\textbf{I}(w^{T}X>0)+\epsilon$$

where $\textbf{I}$ is the indicator function, that is $\textbf{I}(z)$ is unity whenever $z$ is true, and $0$ otherwise.

```python
x_normal, y_normal = sims.step_sim(100, 1)
x_no_noise, y_no_noise = sims.step_sim(1000, 1, noise=0)
```

```python
fig = plt.figure(figsize=(5, 4))
ax1 = fig.add_subplot(1, 1, 1)
ax1.scatter(x_normal, y_normal, c='b', label='Normal (with noise)')
ax1.scatter(x_no_noise, y_no_noise, c='r', label='No noise')
ax1.set_title('Step Simulation', fontweight='bold')
ax1.set_xlabel('X Vector')
ax1.set_ylabel('Y Vector')
plt.legend(bbox_to_anchor=(1.3, 0.9), bbox_transform=plt.gcf().transFigure)
plt.show()
```

## Spiral Simulation

<br></br>
<br></br>
<center>$X_{|d|}=U\sin(\pi U)\cos^{d}(\pi U)$ for $d=1,...,p-1$</center>

$$X_{|p|}=U\cos^{p}(\pi U)$$

$$Y=U\sin(\pi U)+0.4p\epsilon$$

```python
x_normal, y_normal = sims.spiral_sim(100, 1)
x_no_noise, y_no_noise = sims.spiral_sim(1000, 1, noise=0)
```
